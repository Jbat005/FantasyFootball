```python
import pandas as pd
import matplotlib as plt
```

```python
data = pd.read_csv('passing_summary2019.csv')
data
```

```python
player_yards = data.groupby('player').yards.sum().sort_values(ascending=False)


best_players = player_yards.index[:20]
best_yards = data[data.player.isin(best_players)]
best_yards['fantasy_yard_points'] = best_yards.yards / 25 # Passing Yards: Typically, 1 point for every 25 yards
best_yards['fantasy_td_points'] = best_yards.touchdowns * 6 # Touchdowns: 6 points (for passing, rushing, or receiving)
best_yards['fantasy_int_points'] =best_yards.interceptions * -2 # Interceptions: -2 points
#best_yards['fantasy_fumble_points'] = best_yards.fumble * -2 # Fumbles Lost: -2 points
best_yards['fantasy_points'] = best_yards.fantasy_yard_points + best_yards.fantasy_td_points + best_yards.fantasy_int_points # total
best_yards[['player','fantasy_yard_points', 'fantasy_td_points', 'fantasy_int_points','fantasy_points']].sort_values('fantasy_points', ascending=False) 
```

```python
"""Passing Yards: Typically, 1 point for every 25 yards
Rushing Yards: Usually, 1 point for every 10 yards
Receiving Yards: Similar to rushing, 1 point for every 10 yards
Touchdowns: 6 points (for passing, rushing, or receiving)
Interceptions Thrown: -2 points
Fumbles Lost: -2 points
Field Goals: Points vary by distance, e.g., 3 points for under 50 yards
Extra Points: 1 point"""
```
